## 数据准备

### 添加/删除 列(特征)

```python
new_movie_pd = notnull_movie_pd.drop(["时间", "演员"], axis=1)
```

```python
new_movie_pd
```

```python
new_movie_pd["权重评分"] = new_movie_pd["评分"]**(1/2) * new_movie_pd["评价人数"]
```

```python
new_movie_pd
```

```python
def get_level(x):
    if x > 9.0:
        return "A"
    elif x > 7.0:
        return "B"
    else:
        return "C"
new_movie_pd["等级"] = new_movie_pd.评分.map(get_level)
```

```python
new_movie_pd
```

```python
new_movie_pd['等级'] =  pd.cut(new_movie_pd['评分'], bins = [0, 7.0, 9.0, float('Inf')], labels = ['C', 'B', 'A'], right = False)
```

```python
new_movie_pd
```

### 去重

```python
new_movie_pd = new_movie_pd.drop_duplicates(subset=["链接"])
new_movie_pd
```

```python
movie_url = notnull_movie_pd.groupby('链接').size().reset_index(name='count')
movie_url[ movie_url['count'] > 1 ]
```

```python
notnull_movie_pd[notnull_movie_pd.链接 == "https://movie.douban.com/subject/7564985/"]
```

```python
len(new_movie_pd.链接.unique())
```

### 类型转换

```python
new_movie_pd.评分 = new_movie_pd.评分.apply(pd.to_numeric)
new_movie_pd.评价人数 = new_movie_pd.评价人数.astype(int)
```

```python
new_movie_pd.describe()
```

## 可视化

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 3))

new_movie_pd[new_movie_pd.等级 == 'A'].评价人数.plot(ax=axes[0], style="ro", label="A")
new_movie_pd[new_movie_pd.等级 == 'B'].评价人数.plot(ax=axes[1], style="o", label="B")

axes[0].legend()
axes[1].legend()
axes[0].set_xlabel("level: %s" % 'A')
axes[1].set_xlabel("level: %s" % 'B')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 3))

new_movie_pd[new_movie_pd.等级 == 'A'].评价人数.plot(kind="hist", ax=axes[0], color="r", label="A")
new_movie_pd[new_movie_pd.等级 == 'B'].评价人数.plot(kind="hist", ax=axes[1], color="y", label="B")

axes[0].legend()
axes[1].legend()
axes[0].set_xlabel("level: %s" % 'A')
axes[1].set_xlabel("level: %s" % 'B')
```

```python
ax = new_movie_pd.plot(x="评价人数", y="评分", style="ro")
ax.set_xlabel("number of people")
ax.set_ylabel("score")
```

```python
ax = new_movie_pd.plot(x="评价人数", y="评分", style="ro")
ax.set_xlabel("number of people")
ax.set_ylabel("score")
ax.legend(["movie"])
ax.set_xticks(range(0, 1000000, 200000))
ax.axhline(y=6, lw=4, ls="--", color="r")

ax.lines[0].set_color("b")
ax.lines[0].set_marker("s")
```
