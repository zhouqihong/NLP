```python
points = [
    (0, 0), 
    (0.5, 0), 
    (0.5, 6.44),
    (6, 0),
    (6, 8),
    (5.5, 8),
    (5.5, 1.58),
    (0, 8)
]
```

```python
    %matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([0, 1], [1, 2])
```

```python
def draw_segment(segments):
    for i, seg in enumerate(segments):
        _next = segments[(i + 1) % len(segments)]
        x1, y1 = seg
        x2, y2 = _next
        plt.plot([x1, x2], [y1, y2])
```

```python
draw_segment(points)
```

```python
import numpy as np
```

```python
points_matrix = np.array(points)
```

```python
transform = np.array([
    [1, 0.4],
    [0.8, 1],
])
```

```python
tranformed = np.dot(points_matrix, transform)
draw_segment([list(t) for t in tranformed])
```

##  至少出现一次3的概率是多少呢？ 

```python
events = [(a, b) for a, b in occurences if a == 3 or b == 3]
```

```python
def probability(events, occurencces):
    return len(set(events)) / len(set(occurencces))
```

```python
probability(events, occurences)
```

## 如果我第一次出现的是1，那么第二次出现3的概率是多少？

```python
first_is_one = [
    (a, b) for a, b in occurences if a == 1
]
```

```python
probability(second_is_three, first_is_one)
```
