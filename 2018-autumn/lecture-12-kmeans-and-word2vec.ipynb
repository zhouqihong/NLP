```python
all_comments = [' '.join(cut(s)) for s in all_comments if isinstance(s, str)]
```

```python
import random
```

```python
samples = random.sample(all_comments, 10000)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=30000)
```

```python
X = vectorizer.fit_transform(samples)
```

```python
id_word = {i: w for w, i in vectorizer.vocabulary_.items()}
```

```python
def softmax(logits):
    logits = logits - np.max(logits) # normalization
    with_exp = np.exp(logits)
    return with_exp / np.sum(with_exp)
```
